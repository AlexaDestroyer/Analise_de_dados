# Qualidade dos Vinhos üç∑
A an√°lise se refere a Dados dedos Vinhos Tinto e Branco tirado de um dataset p√∫blico,fonte citada c√©lula abaixo, contendo 12 caracter√≠sticas **qu√≠micas** e **sensoriais** para avalia√ß√£o da qualidade de um vinho, s√£o elas:

Caracter√≠sticas **Sensoriais**
**Acidez fixa:** Refere-se √† acidez natural do vinho, proveniente principalmente do √°cido tart√°rico e m√°lico das uvas. N√≠veis mais altos de acidez fixa contribuem para a frescor e vivacidade do vinho, enquanto n√≠veis mais baixos podem resultar em um vinho mais macio e redondo.
**Acidez vol√°til**: √â a parte da acidez do vinho derivada de compostos vol√°teis como √°cido ac√©tico e √°cido etanoico. N√≠veis elevados de acidez vol√°til podem indicar um vinho estragado ou com defeito, enquanto n√≠veis moderados podem agregar complexidade e interesse ao vinho.
**√Åcido c√≠trico**: Um dos principais √°cidos org√¢nicos do vinho, o √°cido c√≠trico contribui para a acidez e o sabor c√≠trico do vinho. N√≠veis equilibrados de √°cido c√≠trico conferem frescor e vivacidade, enquanto n√≠veis excessivos podem resultar em um vinho azedo.
**A√ß√∫car residual**: Refere-se √† quantidade de a√ß√∫car n√£o fermentado que permanece no vinho ap√≥s a fermenta√ß√£o. Vinhos secos t√™m baixo teor de a√ß√∫car residual, enquanto vinhos doces, como os Portos, podem ter um teor de a√ß√∫car residual significante.
**Cloretos**: Presentes em pequenas quantidades no vinho, os cloretos podem contribuir para o sabor salgado ou mineral do vinho. N√≠veis excessivos de cloretos podem indicar um problema de produ√ß√£o.
**Di√≥xido de enxofre livre**: Um conservante adicionado ao vinho para prevenir a oxida√ß√£o e o crescimento microbiano. N√≠veis moderados de di√≥xido de enxofre livre s√£o geralmente aceit√°veis, enquanto n√≠veis excessivos podem ter um sabor amargo ou pungente.
**Di√≥xido de enxofre total**: A soma do di√≥xido de enxofre livre e do di√≥xido de enxofre ligado, este √∫ltimo formado durante a fermenta√ß√£o. N√≠veis mais altos de di√≥xido de enxofre total podem indicar um vinho com maior potencial de oxida√ß√£o.
**Densidade**: Indica a massa do vinho por unidade de volume. Vinhos com maior densidade geralmente t√™m maior teor de √°lcool, extrato e corpo.
**PH:** Uma medida da acidez ou alcalinidade do vinho. N√≠veis de pH mais baixos indicam um vinho mais √°cido, enquanto n√≠veis de pH mais altos indicam um vinho mais alcalino. O pH ideal para o vinho varia de acordo com o estilo e a variedade da uva.
**Sulfatos**: Presentes em pequenas quantidades no vinho, os sulfatos podem contribuir para o sabor salgado ou mineral do vinho. N√≠veis excessivos de sulfatos podem indicar um problema de produ√ß√£o.
**Teor alco√≥lico**: A porcentagem de √°lcool et√≠lico presente no vinho. O teor alco√≥lico varia de acordo com a variedade da uva, o clima e as pr√°ticas de vinifica√ß√£o.

Caracter√≠sticas Sensoriais:
**Qualidade**: Uma avalia√ß√£o geral da qualidade do vinho, levando em considera√ß√£o suas caracter√≠sticas organol√©pticas, como apar√™ncia, aroma, sabor e final de boca.

```python
# Visualiza√ß√£o da matriz de correla√ß√£o de Pearson para VINHO TINTO
corr_tinto = dados_tinto.corr()
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
sns.heatmap(corr_tinto, annot=True, annot_kws={"size": 10}, cbar_kws={"shrink": 0.8}, cmap="coolwarm", fmt=".2f")
plt.title("Qualidade Vinho Tinto", fontsize=14)
plt.xticks(rotation=85, fontsize=8)
plt.yticks(fontsize=8)

# Visualiza√ß√£o da matriz de correla√ß√£o de Pearson para Vinho BRANCO
corr_branco = dados_branco.corr()
plt.subplot(1, 2, 2)
sns.heatmap(corr_branco, annot=True, annot_kws={"size": 10}, cbar_kws={"shrink": 0.8}, cmap="coolwarm", fmt=".2f")
plt.title("Qualidade Vinho Branco", fontsize=14)
plt.xticks(rotation=85, fontsize=8)
plt.yticks(fontsize=8)

plt.tight_layout()
plt.show()

#Correla√ß√£o pr√≥xima de 1, indica uma forte correla√ß√£o positiva, o que significa que quando uma vari√°vel aumenta, a outra tende a aumentar tamb√©m.
#correla√ß√£o √© pr√≥xima de -1, indica uma forte correla√ß√£o negativa, o que significa que quando uma vari√°vel aumenta, a outra tende a diminuir.
#correla√ß√£o √© pr√≥xima de 0, indica que n√£o h√° uma rela√ß√£o linear entre as vari√°veis.

#------------------------------------------CONCLUS√ÉO------------------------------------------------------------------

#No VINHO TINTO temos as correla√ß√µes POSITIVAS entre as MAIORES variaveis POSITIVAS: ENXOFRE LIVRE (O,67) , ENXOFRE TOTAL(0,67), ALCOOL (0,48) E QUALIDADE (0,48)
#No VINHO TINTO temos as correla√ß√µes NEGATIVAS entre as variavies ACIDEZ VOLATIL (-0,55), ACIDO CITRICO (-0,55), DESNSIDADE (-0,34), PH (-0,34)

#No VINHO BRANCO temos as correla√ß√µes POSITIVAS entre as MAIORES variaveis POSITIVAS: ENXOFRE LIVRE (O,62) , ENXOFRE TOTAL(0,53 A 0,62), ALCOOL (0,44) E QUALIDADE (0,44)
##No VINHO BRANCO temos as correla√ß√µes NEGATIVAS entre as MAIORES variavies NEGATIVAS: ACIDEZ VOLATIL (-0,15), ACIDO CITRICO (-0,15), DESNSIDADE (-0,9), PH (-0,9)
```
